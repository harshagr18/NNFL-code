```
import numpy as np
import matplotlib.pyplot as plt
```

```
def identity(x):
    return x

def bin_step(x):
    return 1 * (x >= 0) 

def bipolar_step(x):
  return 1 * (x >= 0) - 1 * (x < 0)

def bin_sigmoidal(x,l):
    return 1 / (1 + np.exp(-l*x))

def bip_sigmoidal(x,l):
    return (1 - np.exp(-l*x)) / (1 + np.exp(-l*x))
```

```
X = np.linspace(-5, 5, 100)
for l in range(1, 5):
  plt.plot(X, bin_sigmoidal(X,l),'b')
  plt.xlabel('X Axis')
  plt.ylabel('Y Axis')
  plt.title('Binary Sigmoid Function')
plt.grid()
plt.text(4, 0.8, r'$binsigmoidal(x)=\frac{1}{1+e^{-x}}$', fontsize=10)
plt.show()

```

```
for l in range(1, 5):
  plt.plot(X, bip_sigmoidal(X,l),'r')
  plt.xlabel('X Axis')
  plt.ylabel('Y Axis')
  plt.title('Bipolar LogSigmoid Function')
plt.grid()
plt.show()

```

```
plt.plot(X, identity(X),'g')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Identity Function')
plt.grid()
plt.show()
```

```
plt.plot(X, bin_step(X),'y')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Binary Step Function')
plt.grid()
plt.show()
```

```
plt.plot(X, bipolar_step(X),'black')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Binary Step Function')
plt.grid()
plt.show()
```
